```
#TRANFORMANDO AS COLUNAS DO TIPO DATA EM DATA.

from pyspark.sql.functions import *
from pyspark.sql.types import DateType
from pyspark.sql.functions import col, unix_timestamp, to_date

df3 = df3.withColumn('data_de_admissao',to_date(unix_timestamp(col('data_de_admissao'), 'dd/MM/yyyy').cast("timestamp")))
df3 = df3.withColumn('data_de_nascimento',to_date(unix_timestamp(col('data_de_nascimento'), 'dd/MM/yyyy').cast("timestamp")))
df3 = df3.withColumn('mes_referencia',to_date(unix_timestamp(col('mes_referencia'), 'dd/MM/yyyy').cast("timestamp")))
```

```
import pyspark  
from pyspark.sql import SparkSession  
from pyspark.sql.functions import *

# Separando os messes para operações e transformando em colunas temporarias.
# aqui iremos pegar os messes da coluna para fazer as operações.

# Mes Refencia
colu = pyspark.sql.functions.split(df3['mes_referencia'], '-')
df3 = df3.withColumn('Mes_refe', colu.getItem(1))

colu1 = pyspark.sql.functions.split(df3['data_de_admissao'], '-')
df3 = df3.withColumn('Mes_adi', colu1.getItem(1))

# Convertendo as strings em integer numeros
from pyspark.sql.types import IntegerType
df3 = df3.withColumn("Mes_refe",
df3["Mes_refe"].cast(IntegerType()))

from pyspark.sql.types import IntegerType
df3 = df3.withColumn("Mes_adi",
df3["Mes_adi"].cast(IntegerType()))
```

```
# fazendo operações entre as colunas criadas
import pyspark.sql.functions as Func
from  pyspark.sql.types import IntegerType
from pyspark.sql.functions import expr
 
df3 = df3.withColumn('Admitido_no_mês',expr("Mes_refe == Mes_adi"))

# Convertendo a Coluna em Tipo Numérico.
from pyspark.sql.types import IntegerType
df3 = df3.withColumn("Admitido_no_mês",
df3["Admitido_no_mês"].cast(IntegerType()))
```

```
# Demitido no mês: essa coluna deverá ser do tipo numérico, respeitando a seguinte condição: se a data de demissão do registro for igual ao mês de #referência analisado, demitido no mês recebe valor é igual a 1, senão recebe valor igual a 0.

# TENTATIVA

# criando a coluna.
df3 = df3.select('*', lit(0).alias("Demitido_no_mês"))

df3 = df3.withColumn('data_de_demissao',to_date(unix_timestamp(col('data_de_demissao'), 'dd/MM/yyyy').cast("timestamp")))

colu2 = pyspark.sql.functions.split(df3['data_de_demissao'], '-')
df3 = df3.withColumn('Mes_demi', colu2.getItem(1))

df3 = df3.withColumn('Demitido_no_mês',expr("Mes_refe == Mes_demi"))

# Convertendo a Coluna em Tipo Numérico.
from pyspark.sql.types import IntegerType
df3 = df3.withColumn("Demitido_no_mês",
df3["Demitido_no_mês"].cast(IntegerType()))
```

```
#3. **Idade**: calcule a idade em anos no mês de referência em relação à data de nascimento.#

# Criando uma nova coluna.
df3 = df3.select('*', lit(0).alias("Idade"))

# TENTATIVA

# criando a coluna temporaria.
colu3 = pyspark.sql.functions.split(df3['data_de_nascimento'], '-')
df3 = df3.withColumn('ano_nasc', colu3.getItem(0))

colu4 = pyspark.sql.functions.split(df3['mes_referencia'], '-')
df3 = df3.withColumn('refe_ano', colu4.getItem(0))

df3 = df3.withColumn('Idade',expr("refe_ano - ano_nasc"))

# Convertendo a Coluna em Tipo Numérico.
from pyspark.sql.types import IntegerType
df3 = df3.withColumn("Idade",
df3["Idade"].cast(IntegerType()))
```
